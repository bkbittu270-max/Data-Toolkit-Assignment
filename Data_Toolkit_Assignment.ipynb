```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
...
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [90, 80, 70, 60, 85],
    'Science': [88, 76, 66, 54, 82],
    'English': [78, 85, 90, 70, 60],
    'History': [60, 65, 70, 72, 68]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

# Calculate correlation matrix
corr_matrix = df.corr()

print("\nCorrelation Matrix:")
print(corr_matrix)

# Plot heatmap using seaborn
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix Heatmap")
plt.show()
...
```

```python
#  Perform matrix multiplication using NumPy.
...
import numpy as np

# Define two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Perform matrix multiplication
result1 = np.dot(A, B)      # Method 1
result2 = np.matmul(A, B)   # Method 2
result3 = A @ B             # Method 3

print("\nResult using np.dot():\n", result1)
print("\nResult using np.matmul():\n", result2)
print("\nResult using @ operator:\n", result3)
...
```
